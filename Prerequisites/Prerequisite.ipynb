# Feature Extraction of Text Data: the mapping from textual data to real valued vectors.
* Bag Of Words (BOW): list of unique words in the text corpus
* Term Frequency - Inverse Document Frequency (TF-IDF): count the number of times each word appears in a document

##
* Term Frequency (TF): number of times term t appears in a document
* Inverse Document Frequency(IDF): log(N/n):
  - N: number of documents
  - n: number of documents a term t has appeared in
  => The IDF value of a rare word is high, the IDF value of a frequent word is low.
TF-IDF value of a term = TF x IDF

## Steming: reducing a word to its Rootword

```python
port_stem = PorterStemmer()
```

```python
def steming(content):
  stemmed_content = re.sub("[^a-zA-Z]", " ", content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words("english")]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
news_data["content"] = steming(news_data["content"].to_string())
```
